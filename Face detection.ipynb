```python
import numpy as np
import cv2
import matplotlib.pyplot as plt

%matplotlib inline
from tkinter import *
from tkinter import messagebox
win=Tk()
win.state("zoomed")
win.resizable(width=False,height=False)
win.configure(bg="black")
win.title("My project")
lbl_title=Label(win,text="FACE DETECTION",font=('',50,'underline'),bg="black",fg="white")
lbl_title.place(relx=.33,rely=0.04)

def home_screen():
    
   
    frm=Frame(win,bg='black')
    frm.place(relx=0,rely=.15,relwidth=1,relheight=1)
    
    

    
    lbl_user=Label(frm,text="-By Vaibhavi srivastava",font=('',20,'bold'),bg='black',fg="white")
    lbl_user.place(relx=.8,rely=.1)
    
    lbl_user=Label(frm,text="Username",font=('',20,'bold'),bg='black',fg="white")
    lbl_user.place(relx=.3,rely=.2)

    entry_user=Entry(frm,font=('',20,'bold'),bd=10)
    entry_user.place(relx=.42,rely=.2)
    entry_user.focus()

    lbl_pass=Label(frm,text="Password",font=('',20,'bold'),bg='black',fg="white")
    lbl_pass.place(relx=.3,rely=.3)

    entry_pass=Entry(frm,font=('',20,'bold'),bd=10,show="*")
    entry_pass.place(relx=.42,rely=.3)
    
    btn_login=Button(frm,command=lambda:welcome_screen(entry_user,entry_pass),text="LOGIN",font=('',20,'bold'),bd=10,width=10)
    btn_login.place(relx=.46,rely=.4)
    
                  

def welcome_screen(entry_user=None,entry_pass=None):
    if(entry_user!=None and entry_pass!=None):
        user=entry_user.get()
        pwd=entry_pass.get()
     
    else:
        user="admin"
        pwd="admin"
    if(len(user)==0 or len(pwd)==0):
        messagebox.showwarning("validation","Please fill the fields")
        return
    else:
        if(user=="admin" and pwd=="admin"):
            
            frm=Frame(win,bg='Red')
            frm.place(relx=0,rely=0,relwidth=1,relheight=1)
            lbl_title=Label(win,text="MAIN MENU",font=('',50,'underline'),bg="black",fg="white")
            lbl_title.place(relx=.39,rely=0.15)
            
            btn_l=Button(frm,command=lambda:video(entry_user),text="USING VIDEO",font=('',20,'bold'),bd=10,width=20)
            btn_l.place(relx=.40,rely=.5)
            btn_l=Button(frm,command=lambda:webcam(entry_user),text="USING WEBCAM",font=('',20,'bold'),bd=10,width=20)
            btn_l.place(relx=.40,rely=.6)
        else:
            messagebox.showerror("Fail","Invalid username/password")
        btn_log=Button(frm,command=lambda:Exit(),text="LOGOUT",font=('',20,'bold'),bd=10,width=10)
        btn_log.place(relx=.46,rely=.7)
            
            
            


    
    
    
'''-------------------------------video------------------------------------------------------------------------------------------'''
def video(entry_user):
    frm=Frame(win,bg='white')
    frm.place(relx=0,rely=0,relwidth=1,relheight=1)
    lbl_title=Label(win,text="USING VIDEO",font=('',20,'underline'),bg="black",fg="white")
    lbl_title.place(relx=.45,rely=0.04)
    btn_log=Button(frm,command=lambda:video2(entry_user),text="Start",font=('',20,'bold'),bd=10,width=24)
    btn_log.place(relx=.37,rely=.4)
    
    lbl_user=Label(frm,text="-Pause:Press P ",font=('',10,'bold'),bg="white",fg="black")
    lbl_user.place(relx=.89,rely=.7)
    lbl_user=Label(frm,text="-Resume:Press any key ",font=('',10,'bold'),bg="white",fg="black")
    lbl_user.place(relx=.89,rely=.8)
    lbl_user=Label(frm,text="-Stop:Press S ",font=('',10,'bold'),bg="white",fg="black")
    lbl_user.place(relx=.89,rely=.9)
    btn_log=Button(frm,command=lambda:Back(),text="Go back to main menu",font=('',20,'bold'),bd=10,width=24)
    btn_log.place(relx=.37,rely=.5)
    btn_log=Button(frm,command=lambda:Exit(),text="LOGOUT",font=('',20,'bold'),bd=10,width=10)
    btn_log.place(relx=.45,rely=.6)
def video2(entry_user):
    messagebox.showinfo("Information","Pause: Press P\nResume: Press any key\nStop: Press S",)
    face_cascade = cv2.CascadeClassifier('C:/Users/vaibh/Downloads/haarcascade_frontalface_default.xml')  
    cap = cv2.VideoCapture('C:/Users/vaibh/Downloads/videoplayback.mp4')
    
    
    while(True):
                flag,img=cap.read()
                if(flag==True):
                    gray=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
                    faces=face_cascade.detectMultiScale(gray,1.4,4)
                    for x,y,w,h in faces:
                        cv2.rectangle(img,(x,y),(x+w,y+h),(0,0,255),2)
                    cv2.imshow('cap',img)
                    k=cv2.waitKey(1)
                    if (k == ord('p')):
                        cv2.waitKey()
                    elif (k == ord('s')):
                        break  
                   
    cap.release()
    cv2.destroyAllWindows()
    
    
    
    
    
'''----------------------------webcam---------------------------------------------------------------------------------------------'''  
    
def webcam(entry_user):
    frm=Frame(win,bg='white')
    frm.place(relx=0,rely=0,relwidth=1,relheight=1)
    lbl_title=Label(win,text="USING WEBCAM",font=('',20,'underline'),bg="black",fg="white")
    lbl_title.place(relx=.44,rely=0.04)
    frm.place(relx=0,rely=0,relwidth=1,relheight=1)
    btn_log=Button(frm,command=lambda:webcam2(entry_user),text="Turn on the camera",font=('',20,'bold'),bd=10,width=24)
    btn_log.place(relx=.37,rely=.4)
    lbl_user=Label(frm,text="-To turn off the camera: Press S",font=('',20,'bold'),bg='white',fg="black")
    lbl_user.place(relx=.7,rely=.9)
    btn_log=Button(frm,command=lambda:Back(),text="Go back to main menu",font=('',20,'bold'),bd=10,width=24)
    btn_log.place(relx=.37,rely=.5)
    btn_log=Button(frm,command=lambda:Exit(),text="LOGOUT",font=('',20,'bold'),bd=10,width=10)
    btn_log.place(relx=.45,rely=.6)
    
def webcam2(entry_user):
    messagebox.showinfo("Information","To turn off the camera: Press S",)
    face_cascade = cv2.CascadeClassifier('C:/Users/vaibh/Downloads/haarcascade_frontalface_default.xml')
    video = cv2.VideoCapture(0)

    while True:
        check, frame = video.read();
        faces = face_cascade.detectMultiScale(frame,scaleFactor=1.1, minNeighbors=5);
        for x,y,w,h in faces:
            frame = cv2.rectangle(frame, (x,y), (x+w,y+h), (0,255,0), 3)
        cv2.imshow('Face Detector', frame)
        key = cv2.waitKey(1)
        if key == ord('s'):
            break;

    video.release()
    cv2.destroyAllWindows()
    
    
'''-----------------------back--------------------------------------------------------'''
def Back():
    option=messagebox.askyesno('Confirmation','Are you sure you want to go Back?')
    if(option==True):
        
        welcome_screen()
    else:
        pass
    
'''----------------------------------------exit-----------------------------------------------------------------------------------'''          
def Exit():
    
    option=messagebox.askyesno('Confirmation','Are you sure you want to exit?')
    if(option==True):
        frm=Frame(win,bg='black')
        frm.place(relx=0,rely=0,relwidth=1,relheight=1)
        
        lbl_title=Label(win,text="FACE DETECTION",font=('',50,'underline'),bg="black",fg="white")
        lbl_title.place(relx=.33,rely=0.04)
        home_screen()
    else:
        pass
    
    
home_screen()             
win.mainloop()
```
